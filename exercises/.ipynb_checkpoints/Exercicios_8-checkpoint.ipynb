<a id='funcao1'></a>

<h3>2.2 Função cálculo do k-means</h3>

A função abaixo recebe como parâmetros a quantidade de <i>clusters</i> que o k-means deverá encontrar. O parâmetro nRodadas define quantas vezes o algoritmo irá rodar. Por fim, <i>centers_param</i> indica o método que será utilizado para a inicialização das posições dos centróides.

A primeira etapa é a conversão da imagem em um vetor. Em seguida, define-se o critério de parada do algortimo, sendo que as condições definidas foram a quantidade máxima de iterações ou uma tolerância mínima. A função <i>cv2.kmeans</i> irá retornar os reotulos encontrados e as posições dos centros. 

Em seguida, realiza-se o processo inverso, de conversão dos vetores resultantes do k-means para uma matriz correspondendo a uma imagem, sendo esta retornada.

```python
def calc_kmeans(nClusters, nRodadas, centers_param):
    samples = np.zeros([img.shape[0]*img.shape[1],3], dtype='float32')
    
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            for z in range(3):
                samples[y + x*img.shape[0],z] = img[y,x,z]
    
    criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_MAX_ITER, 10000, 0.0001)
    
    ret, rotulos, centros = cv2.kmeans(samples, nClusters, None, criteria, nRodadas, centers_param)
    
    rotulada = np.zeros(img.shape, dtype='uint8')
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            indice = rotulos[y + x*img.shape[0],0]
            rotulada[y,x,0] = int(centros[indice,0])
            rotulada[y,x,1] = int(centros[indice,1])
            rotulada[y,x,2] = int(centros[indice,2])
    
    return rotulada
```

Para o presente exercício,  a quantidade de <i>clusters</i> definida foram 8 e a quantidade de rodadas do algoritmo foi definida como uma, como mostrado abaixo:

```python
nClusters = 8
nRodadas = 1

img = cv2.imread("..\\images\\sushi.jpg")
```

<a id='result1'></a>

<h3>2.3 Resultados</h3>

<p style='text-align: justify;'>No trecho de código abaixo, o k-means é rodado 10 vezes sobre uma mesma imagem, de forma que o parâmetro de inicialização dos centros é o <i>cv2.KMEANS_RANDOM_CENTERS</i>, ou seja, os primeiros pontos são definidos de forma aleatória. Em seguida, todas as imagens resultantes do processo foram exibidas. </p>

```python
images = []
for i in range(10):
    images.append(calc_kmeans(8, nRodadas, cv2.KMEANS_RANDOM_CENTERS))

plt.figure(1,[15,20])

for i in range(10):
    plt.subplot(5,2,i+1)
    plt.imshow(cv2.cvtColor(images[i],cv2.COLOR_BGR2RGB))
    plt.title("Rodada "+str(i+1))
```

Apesar de sutil, percebe-se que existem diferenças entre todas as dez imagens geradas. Isto se deve pelo fato de o algoritmo ser não determinístico, ou seja, para uma mesma entrada no cálculo do k-means, diferentes resultados podem ser gerados. Outro ponto que corrobora com isto é a inicialização dos centros em pontos aleatórios que, durante as iterações do cálculo, tenderão a centros diferentes no final do processo.
